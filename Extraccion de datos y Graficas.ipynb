```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Data.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
# Rename the columns for easier access
df.columns = ['timestamp', 'thingName', 'Peso']

# Convert the 'timestamp' column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Set the 'timestamp' column as the index
df.set_index('timestamp', inplace=True)
```

# GRAFICA EVOLUCION DE LA BASURA EN EL TIEMPO

```python
#Tamao de la figura
plt.figure(figsize=(12, 6))

#Grafica de lineas
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Peso'], marker='o', linestyle='-', color='b')
plt.title('Peso over Time')
plt.xlabel('Timestamp')
plt.ylabel('Peso')
plt.grid(True)
plt.show()

#Grfica de barras
plt.bar(df.index, df['Peso'], color='green')
plt.xlabel('Fecha')
plt.ylabel('Peso')
plt.xticks(rotation=45)
plt.grid()
plt.title('Evolucin del Peso en el Tiempo')

plt.show()
```

# GRAFICA DE LA BASURA POR DIA

```python
peso_diario = df['Peso'].resample('D').sum()
```

```python
# Configurar el tamao de la figura
plt.figure(figsize=(12, 6))

# Crear la grfica de barras
peso_diario.plot(kind='bar', figsize=(16, 6))

# Aadir etiquetas a los ejes
plt.xlabel('Fecha')
plt.ylabel('Peso Total')
plt.title('Peso Total por Da')

# Rotar las etiquetas del eje x para evitar que se superpongan
plt.xticks(rotation=45)

# Mostrar la grfica
plt.show()
```

# GRAFICA DE LA BASURA POR SEMANA

```python
peso_semanal = df['Peso'].resample('W').sum()
```

```python
plt.figure(figsize=(12, 6))
peso_semanal.plot(kind='bar', figsize=(16, 6))
plt.xlabel('Semana')
plt.ylabel('Peso Total')
plt.title('Peso Total por Semana')
plt.xticks(rotation=45)
plt.show()
```
